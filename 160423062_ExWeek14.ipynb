```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("births.csv")
print(df)

feb_df = df[df['month'] == 2]

grouped = feb_df.groupby(['year', 'gender'])['births'].sum().reset_index()

pivot = grouped.pivot(index='year', columns='gender', values='births')

plt.figure(figsize=(10,6))
plt.plot(pivot.index, pivot['M'], marker='o', color='orangered', label='Male')
plt.plot(pivot.index, pivot['F'], marker='o', color='gold', label='Female')

plt.title("Total Births in February by Gender (19692008)")
plt.xlabel("Year")
plt.ylabel("Total Births in February")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
np.random.seed(0)

hours = np.arange(6,18)

temperatures = np.random.normal(loc=28, scale=2, size=len(hours))

for i, txt in enumerate(temperatures):
    plt.annotate(np.round(txt,2), (hours[i], temperatures[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.plot(hours,temperatures, color="r", linestyle="-", marker="o")
plt.Figure(figsize=(6,8))
plt.xlabel("Jam (WIB)")
plt.ylabel("Suhu(C)")
plt.title("Tren Suhu Ruangan per Jam")
plt.grid(1)
plt.show()
```
