```python
import numpy as np
```

```python
p = int(input("Masukkan jumlah produk: "))
k = int(input("Masukkan jumlah kriteria penilaian: "))

arrRating = np.random.randint(1,6, size=(p,k))
print("Isi arrRating:\n",arrRating)

weightOfCriteria = 10 *(np.random.randint(1,11, size=(1,k)))
print("Weight Of Criteria: ",weightOfCriteria)
totalWeight = np.sum(weightOfCriteria)
print("Total weight: ",totalWeight)
normalizeWOC = np.round(weightOfCriteria/totalWeight, 2)
print("Normalize WOC: ",normalizeWOC)
print("Normalize WOC size: ",normalizeWOC.size)

finalRating = np.sum(np.array(arrRating*normalizeWOC), axis=1)
print("Final rating before reshape: ",finalRating)
finalRating = finalRating.reshape(p,1)
print("Final rating after reshape: \n",finalRating)
print("Final rating size: ",finalRating.size)

productRating = np.concatenate([arrRating, finalRating], axis=1)
averageRating = np.average(productRating, axis=0)
print("Rata-rata kolom: \n", averageRating)
print("Jumlah produk yang ratingnya di bawah 3.0:\n", np.sum(productRating < 3.0))

```

```python
p = int(input("Jumlah produk yang ingin dianalisis: "))
k = int(input("Jumlah kategori pilihan: "))

arrProducts = np.random.randint(10,101, size=(p,k))
print("arrProducts: \n",arrProducts)
print("=========================")
print("Jumlah Dimensi: ", arrProducts.ndim)
print("Ukuran: ", arrProducts.shape)
print("Bentuk: ", arrProducts.size)

print("\nBAGIAN 2\n")

print("Jumlah elemen arrProducts yang di atas 80: ", np.sum(arrProducts>80))
arrAbove50 = arrProducts > 50
print("arrAbove50: \n", arrAbove50)

print("\nBAGIAN 3\n")

rataRataNilaiProduk = np.mean(arrProducts, axis=1)
print("Rata-rata Nilai produk: \n", rataRataNilaiProduk)

def is_promo(price):  
    return price < 50

arrayPromo = is_promo(rataRataNilaiProduk)
print("arrayPromo: \n", arrayPromo)

produkLayakDiskon = np.sum(arrayPromo == 1)

# atau bisa pakai
# produkLayakDiskon = np.sum(arrayPromo) --> katanya lebih pythonic
print("Jumlah produk layak diskon:",produkLayakDiskon)

print("\nBAGIAN 4\n")
urutArrayProducts = np.sort(arrProducts, axis=1)
print("Array Products yang urut: \n", urutArrayProducts)

produkRataRataTertinggi = np.sort(rataRataNilaiProduk)
print(produkRataRataTertinggi)
print("3 produk dengan rata-rata kategori tertinggi:", produkRataRataTertinggi[-3:])

isiKolomIndex1 = arrProducts[:, 1:2]
print("isiKolomIndex1:",isiKolomIndex1)

isiKolomIndex3 = arrProducts[:, 3:4]
print("isiKolomIndex3:",isiKolomIndex3)

print("\nBAGIAN 5\n")

averageScore = np.mean(arrProducts, axis=1).reshape(p,1)
print("Average Score:\n",averageScore)

fullReport = np.concatenate([arrProducts,averageScore], axis=1)
print("FULL REPORT: \n",fullReport)

print("Jumlah produk yang rata-rata di atas 75:", np.sum(averageScore>75))

# Ambil indeks urutan berdasarkan kolom terakhir (average score)
sorted_indices = np.argsort(fullReport[:, -1])[::-1]  # descending

# Gunakan fancy indexing untuk mengurutkan fullReport
sorted_fullReport = fullReport[sorted_indices]

# Tampilkan hasilnya
print(sorted_fullReport)
```

```python
np.random.seed(0)

print("\nBAGIAN 1\n")

suhuMentah = np.random.randint(20,40, size=(6,8))
print("SUHU MENTAH: \n",suhuMentah)
print("Suhu Sensor pada 5 jam pertama: ")
print("Suhu sensor ke-2:\n", suhuMentah[:5, [1]])
print("Suhu sensor ke-4:\n", suhuMentah[:5, [3]])

suhu3D = suhuMentah.reshape(3,4,4)
print("SUHU 3D:\n",suhu3D)
print("Dimensi suhu3D:",suhu3D.ndim)

print("\nBAGIAN 2\n")
print("vsplit:\n",np.vsplit(suhuMentah,2))

hasilVStack = np.vstack(np.vsplit(suhuMentah,2))
print("vstack:\n",hasilVStack)

# harus reshape dulu
statusSensor = np.array([1, 0, 1, 1, 0, 1]).reshape(6, 1)
print("hstack:\n",np.hstack([statusSensor, hasilVStack]))

print("\nBAGIAN 3\n")

def status_panas(suhu):
    return "PANAS" if suhu > 35 else "NORMAL"

statusUFunc = np.frompyfunc(status_panas, 1, 1) #hanya bisa menerima 1 input dan 1 output
hasil = statusUFunc(suhuMentah.flatten())
print(hasil)

print("Jumlah data yang 'PANAS':", np.sum(hasil == "PANAS"))

print("\n BAGIAN 4\n")

print("Posisi suhu yang tepat bernilai 30:", np.where(suhuMentah == 30))

suhu1D = suhuMentah.flatten()
print("Suhu 1D:\n",suhu1D)

hasilPartition = np.partition(suhu1D,5)
print("Hasil Partition:",hasilPartition)

nilaiTerendah5 = hasilPartition[:5]
print("5 Nilai terendah:",nilaiTerendah5)

suhu1D = np.sort(suhuMentah.flatten())
posisi = np.searchsorted(suhu1D, 31.5)
print("Suhu 31.5 seharusnya dimasukkan di indeks ke-", posisi)
```
