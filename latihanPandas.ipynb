```python
print("\n=====3=====\n")
# untuk assign Not an Number bisa pakai np.nan
harga_lama = pd.Series([10000, np.nan, 12000, 14000])
harga_baru = pd.Series([12000, 13000, np.nan, np.nan])

data_pembagian = harga_baru/harga_lama
print("BAGI:\n", data_pembagian)
print("Fill NaN:\n", data_pembagian.fillna(1)) #mengisi data yang NaN dengan angka 1

data_penjualan = pd.DataFrame({
    'Produk': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Speaker', 'Webcam'],
    'Wilayah': ['Jakarta', 'Surabaya', 'Bandung', 'Jakarta', 'Surabaya', 'Bandung'],
    'Penjualan_Q1': [100, 150, np.nan, 120, 80, 70],
    'Penjualan_Q2': [110, np.nan, 90, 130, np.nan, 75],
    'Penjualan_Q3': [np.nan, 160, 95, 125, 85, np.nan],
    'Penjualan_Q4': [130, 170, 100, np.nan, 90, 80]
})
print(data_penjualan,"\n")
print(data_penjualan.isnull()) #cek pada setiap kolom, apakah ada kolom yang datanya bernilai null? jika iya return true
print("Hapus data penjualan yang di barisnya ada NaN\n", data_penjualan.dropna(axis=1, how="any")) #jika ada 1 data yg NaN, maka lgsg drop
print("Hapus data penjualan yang semua data di kolomnya bernilai NaN:\n", data_penjualan.dropna(axis=0, how="all")) #jika semua data pada suatu kolom atau baris NaN, baru drop

print("SOAL TERAKHIR:")
print("ISI SEMUA NILAI NAN DENGAN RATA-RATA DARI KOLOM MASING-MASING")

# Daftar kolom penjualan yang ingin diisi NaN-nya
kolom_penjualan = ['Penjualan_Q1', 'Penjualan_Q2', 'Penjualan_Q3', 'Penjualan_Q4']

# Hitung rata-rata hanya untuk kolom-kolom penjualan yang spesifik
# .mean() akan secara otomatis mengabaikan NaN saat menghitung rata-rata
rata_rata_kolom_penjualan = data_penjualan[kolom_penjualan].mean()

# Buat salinan DataFrame untuk menghindari SettingWithCopyWarning
isi_data_penjualan = data_penjualan.copy()

# Isi NaN hanya pada kolom-kolom penjualan yang spesifik
isi_data_penjualan[kolom_penjualan] = isi_data_penjualan[kolom_penjualan].fillna(rata_rata_kolom_penjualan)

print(isi_data_penjualan)
```

```python
print("\n=====4=====\n")
df_produk_elektronik = pd.DataFrame({   
    "id-produk":[0,1,2], 
    "nama":["setrika", "mic", "oven"],
    "harga":[10000, 15000, 20000]
}, index=[0,1,2])

df_produk_pakaian = pd.DataFrame({  
    "id-produk":[1,3,4],                                    
    "nama":["kaos", "celana", "sarung"],
    "harga":[5000, 2000, 1000]
}, index=[1,3,4])

print("Concat axis = 0\n",pd.concat([df_produk_elektronik, df_produk_pakaian], axis=0))
print("\nConcat axis = 1\n",pd.concat([df_produk_elektronik, df_produk_pakaian], axis=1))

# memastikan apakah saat menggabungkan 2 list (baik series maupun df) ada key yg terduplicate atau tidak --> pakai parameter verify_integrity=True
try:
    print("Concat axis = 0\n",pd.concat([df_produk_elektronik, df_produk_pakaian], axis=0, verify_integrity=True)) #hasilnya return exception
except ValueError as e:
    print("ValueError:",e)
    
# membiarkan key value yang terduplicate --> ya lgsg concate aja gapapak kok
print("Ignore_index:\n", pd.concat([df_produk_elektronik, df_produk_pakaian], axis=1, ignore_index=1))

df_inventori = pd.DataFrame({
    "id-produk":[1,2,5],
    "stok":[10,20,30]
}, index=[0,1,2])

#default dari merge join itu inner, define parameter join yg mau digunakan pakai how=""
# inner join biasa dipakai merge bukan concat
print("Merge Inner Join:\n", pd.merge(df_produk_elektronik, df_inventori, on="id-produk", how="inner"))
print("Merge Outer Join:\n", pd.merge(df_produk_elektronik, df_inventori, on="id-produk", how="outer"))

df_karyawan_info = pd.DataFrame({
    "id":[0,1,2],
    "nama":['adi', 'budi', 'chika'],
    "jabatan":['CEO', 'supervisor', 'karyawan']
}, columns=["id", "nama", "jabatan"])

df_gaji_karyawan = pd.DataFrame({
    "karyawanID":[0,1,2],
    "gaji pokok":[10000, 8000, 2000]
}, columns=["karyawanID", "gaji pokok"])

print("Merge menggunakan left_on dan right on:\n",
      pd.merge(df_karyawan_info, df_gaji_karyawan, left_on="id", right_on="karyawanID",).drop("karyawanID", axis=1))
```

```python
print("\n=====5=====\n")
data_klub = pd.DataFrame({
    "Nama Klub": [
        'Real Madrid', 'Barcelona', 'Bayern Munich', 'Liverpool', 'Man City',
        'Real Madrid', 'Barcelona', 'Bayern Munich', 'Liverpool', 'Man City',
        'Real Madrid', 'Barcelona', 'Bayern Munich', 'Liverpool', 'Man City',
        'Juventus', 'Inter Milan', 'AC Milan', 'Napoli', 'AS Roma',
        'Juventus', 'Inter Milan', 'AC Milan', 'Napoli', 'AS Roma'
    ],
    "Tahun": [
        2022, 2022, 2022, 2022, 2022,
        2023, 2023, 2023, 2023, 2023,
        2024, 2024, 2024, 2024, 2024,
        2022, 2022, 2022, 2022, 2022,
        2023, 2023, 2023, 2023, 2023
    ],
    "Liga": [
        'La Liga', 'La Liga', 'Bundesliga', 'Premier League', 'Premier League',
        'La Liga', 'La Liga', 'Bundesliga', 'Premier League', 'Premier League',
        'La Liga', 'La Liga', 'Bundesliga', 'Premier League', 'Premier League',
        'Serie A', 'Serie A', 'Serie A', 'Serie A', 'Serie A',
        'Serie A', 'Serie A', 'Serie A', 'Serie A', 'Serie A'
    ],
    "Poin": [
        85, 88, 78, 92, 95,  # 2022
        82, 90, 80, 89, 93,  # 2023
        87, 85, 79, 90, 94,  # 2024
        75, 78, 70, 68, 65,  # 2022
        77, 80, 72, 70, 67   # 2023
    ]
})

grouped = data_klub.groupby("Tahun")
for i,j in grouped:
    print(i)
    print(j)

group_nama_klub = data_klub.groupby("Nama Klub")

point_klub = group_nama_klub["Poin"].sum()
print("\n\nTotal poin yang diperoleh:\n",point_klub)

# Untuk mendapatkan poin maksimum setiap klub per tahun, grouping harus pada kedua kolom tersebut
poin_maks_klub_tahun = data_klub.groupby(['Nama Klub', 'Tahun'])['Poin'].max()
print("\n\nPoin maksimum yang diperoleh setiap 'Klub' dalam setiap 'Tahun' (Perbaikan):\n", poin_maks_klub_tahun)

print("\n\nData Klub berdasarkan poin (descanding):\n", data_klub.sort_values(by="Poin", ascending=False))
```

```python
print("=====6=====")
print("\nPivot table yang menunjukkan rata-rata 'Poin' untuk setiap 'Nama Klub' berdasarkan 'Tahun':\n",
      data_klub.pivot_table(
          columns="Tahun", values="Poin", index="Nama Klub", aggfunc={"Poin":"mean"}, fill_value="0.0"
      ))
print("\n\nPivot table yang menunjukkan total 'Poin' untuk setiap 'Liga' berdasarkan 'Tahun'.:\n",
      data_klub.pivot_table(aggfunc={"Poin":"sum"}, index="Liga", columns="Tahun"))
```

SUMMARY (Pivot Table):
- Untuk Setiap --> Index
- Berdasarkan --> Columns
- Menunjukkan --> aggrFunc

```python
# Buat ulang daftar_email untuk demonstrasi berurutan
daftar_email_modifikasi = pd.Series(['Andi@example.com', 'budi.santoso@domain.co.id', 'citra@mail.net', 'dewi@gmail.com'])
print("Series Awal:\n", daftar_email_modifikasi)

# Operasi 1: lower()
daftar_email_modifikasi = daftar_email_modifikasi.str.lower()
print("\nSetelah lower():\n", daftar_email_modifikasi)

# Operasi 2 (walaupun ekstrak tidak mengubah Series): nama pengguna
nama_pengguna_saja = daftar_email_modifikasi.str.split("@").str[0]
print("\nNama Pengguna Saja (tidak mengubah Series):\n", nama_pengguna_saja)

# Operasi 3 (walaupun ekstrak tidak mengubah Series): domain
domain_saja = daftar_email_modifikasi.str.split("@").str[1]
print("\nDomain Saja (tidak mengubah Series):\n", domain_saja)

# Operasi 4 (walaupun contains tidak mengubah Series): cek .com
mengandung_com = daftar_email_modifikasi.str.contains("\\.com", regex=True)
print("\nMengandung '.com' (tidak mengubah Series):\n", mengandung_com)

# Operasi 5: replace()
daftar_email_modifikasi = daftar_email_modifikasi.str.replace("\\.com", ".org", regex=True)
print("\nSetelah replace '.com' dengan '.org':\n", daftar_email_modifikasi)
```
